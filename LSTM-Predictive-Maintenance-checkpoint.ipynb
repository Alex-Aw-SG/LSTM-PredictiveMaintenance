```python
 ## task 3
 ## rul remaining life DF 
 ## rul_df is new with 2 cols, first col id and second col = cycle max col
rul_df  = pd.DataFrame(train_ds.groupby('id')['cycle'].max()).reset_index() 

rul_df.columns = ['id', 'max']
rul_df
```

```python
## train ds merge with the above DF left join based on unique ID , engine id 

train_ds = train_ds.merge(rul_df, on = ['id'], how='left')
## creating new column rul  =  max- cycle
train_ds['rul'] = train_ds['max'] - train_ds['cycle']
# drop max 
train_ds.drop('max', axis= 1, inplace=True)
train_ds.head()
```

```python
# task 3 
## generating labels columns for training data 

w1 = 30 
w0 = 15 
# if rul < 30 , label1  1 or 0
# label1 is new columns with values 0, 1 
train_ds['label1'] =  np.where(train_ds['rul'] <= w1, 1, 0)
#train_ds['label2'] = 

# label2 is new columns with values 2
train_ds['label2'] = train_ds['label1']

# label2 also shoul dshow 0,1 unless untill if its values falls hort of 15 , if <= 15 then label 2 must show 2
# if rul <= 15  , label2   = 2 
train_ds.loc[train_ds['rul'] <= w0 , 'label2' ] = 2 
train_ds.head()
```

```python
# function creation
# generate the sequnces
def generate_sequence(samples, timesteps, no_of_cols):
  ### only sequences which are considered for window length(50)
  ## for one if we put all the tows in a single matrix
  ## samples. -dateframe [no of cols in that df] eg: 25 columns
  ## no_of_elements. = all the rows
  ## time-steps= window size = 50
  single_array = samples[no_of_cols].values
  no_of_elements = single_array.shape[0]
  ## loop
  ## eg id 1 have 192 rows and timesteps = 50
  ### 0-50- row 0-50 
  ### 1     row 1-51
  ### 0-2   row 2-52

  for start,stop in zip(range(0, no_of_elements-timesteps), range(timesteps, no_of_elements)):
    yield single_array[start:stop,:]
```

```python
# pick the feature columns
# all those columns which are starting with s
sensor_cols = []
sensor_cols = ['s' + str(i) for i in range(1,22)]
sequence_cols = []
sequence_cols = ['setting1', 'setting2', 'setting3', 'cycle_norm']
sequence_cols.extend(sensor_cols)
```

```python
# create a generator for the sequences
# calling the generate_sequence to generate the sequence
seq_gen = (list(generate_sequence(train_ds[train_ds['id']==id], window_cycle_size, sequence_cols))
          for id in train_ds['id'].unique())
```

```python
# generate seuqnces and convert to numpy array
## 3d array
seq_array = np.concatenate(list(seq_gen)).astype(np.float32)
seq_array.shape

# 15631 = data (rows)
# 50 = window size
# 25 = features

# shape
#(15631,50,25)
```

```python
# function creation
# y labels or predicted labels
# generate the labels
def generate_labels(samples, timesteps, no_of_cols):
  ### only sequences which are considered for window length(50)
  ## for one if we put all the tows in a single matrix
  single_array = samples[no_of_cols].values
  ## shape[0] = no of rows
  no_of_elements = single_array.shape[0]
  ## loop
  ## eg id 1 have 192 rows and timesteps = 50
  ### 0-50- row 0-50
  ### 1     row 1-51
  ### 0-2 row 2-52
  return single_array[timesteps: no_of_elements,:]
```

```python
# generate labels = predicted col

label_gen = [generate_labels(train_ds[train_ds['id']==id], window_cycle_size, ['label1'])
            for id in train_ds['id'].unique()]

## expected output
### (15631,1)
```

```python
# generate sequences and convert to numpy array
## 3d array
label_predicted_array = np.concatenate(list(label_gen)).astype(np.float32)
label_predicted_array.shape
```

## Create LSTM Model

```python
## LSTM steps
## 1. sequential
## 2. add(LSTM())
## 3. compile()
## 4. fit()
```

```python
# features
# window size
# LSTM
# 1
# 25 input features
no_of_features = seq_array.shape[2]
## 1 output features / field = label1
no_of_output_features = label_predicted_array.shape[1]

model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape = (window_cycle_size, no_of_features)))
model.add(Dropout(0.2))
model.add(LSTM(100, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units = no_of_output_features, activation='sigmoid'))
```

```python
## Compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
model.fit(seq_array, label_predicted_array, batch_size=200, epochs=100, verbose=1,
          validation_split = 0.05,)
```

```python
scores = model.evaluate(seq_array, label_predicted_array, verbose=1, batch_size=200)
```
